### Check Outlier ###

```python
# Selecting numerical columns
num_outlies = data.select_dtypes(include=['int64', 'float64'])
num_columns = num_outlies.shape[1]  # Total number of numerical columns

# Define subplot layout (3 in a row)
cols_per_row = 3
rows = int(np.ceil(num_columns / cols_per_row))  # Calculate required rows

# Create subplots
fig, axes = plt.subplots(rows, cols_per_row, figsize=(18, rows * 5))  

# Flatten the axes array for easy iteration
axes = axes.flatten()  

# Loop through each numerical column and create a boxplot
for i, col in enumerate(num_outlies.columns):
    axes[i].boxplot(num_outlies[col])  
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)  
    axes[i].set_ylabel('Values')  
    axes[i].grid(True)

# Hide any extra subplots if total columns < (rows * cols_per_row)
for j in range(i + 1, len(axes)):  
    fig.delaxes(axes[j])  

# Adjust layout to prevent overlapping
plt.tight_layout()  

# Show the plot
plt.show()
```

```python
num_col = data.select_dtypes(include=['int64', 'float64']).columns

def remove_outlliers(data,col):
    for col in num_col:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        
        IQR = Q3-Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
        
    return data


clean_df = remove_outlliers(data, num_col)


print("Before removing outliers: ", data.shape)
print("After removing outliers: ", clean_df.shape)
```

```python
# Selecting numerical columns
numerical_columns = clean_df.select_dtypes(include=['int64', 'float64'])
num_columns = numerical_columns.shape[1]  # Total number of numerical columns

# Define subplot layout (3 in a row)
cols_per_row = 3
rows = int(np.ceil(num_columns / cols_per_row))  # Calculate required rows

# Create subplots
fig, axes = plt.subplots(rows, cols_per_row, figsize=(18, rows * 5))  

# Flatten the axes array for easy iteration
axes = axes.flatten()  

# Loop through each numerical column and create a boxplot
for i, col in enumerate(numerical_columns.columns):
    axes[i].boxplot(numerical_columns[col])  
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)  
    axes[i].set_ylabel('Values')  
    axes[i].grid(True)

# Hide any extra subplots if total columns < (rows * cols_per_row)
for j in range(i + 1, len(axes)):  
    fig.delaxes(axes[j])  

# Adjust layout to prevent overlapping
plt.tight_layout()  

# Show the plot
plt.show()
```
